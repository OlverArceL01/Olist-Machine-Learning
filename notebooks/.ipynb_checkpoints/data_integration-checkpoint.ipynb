# Data Integration

This process merges information from different tables `orders`, `order_items`, `products`, `customers`, and `product_category_name_translation` to build a single integrated dataset.

## Load Olist Tables

```python
import pandas as pd
from pathlib import Path
from omegaconf import OmegaConf
from olist_ml.config import OlistConfig
from sqlalchemy import create_engine
root_path = project_root=Path.cwd().parent
cfg = OmegaConf.load("../conf/config.yaml")

raw_data_path = root_path / cfg.paths.raw_data
olist_sqlite_path = raw_data_path / cfg.files.olist_sqlite
db_file = f"sqlite:///{olist_sqlite_path}"
engine = create_engine(db_file)

orders_df = pd.read_sql("SELECT * FROM orders;", engine)
order_items_df = pd.read_sql("SELECT * FROM order_items;", engine)
products_df = pd.read_sql("SELECT * FROM products;", engine)
customers_df = pd.read_sql("SELECT * FROM customers;", engine)
product_category_name_translation_df = pd.read_sql("SELECT * FROM product_category_name_translation;", engine)
```

## Dataset Granularity
It's necessary to consider the two main problems addressed in this project:
1. `Demand Forecasting`: Based on historical sales data, how much demand can we expect for each product in the coming weeks or months?.
2. `Customer Segmentation`: What types of customers can be identified according to their purchasing behavior?.

Therefore, the dataset granularity is defined at the order item level, as this level provides not only information about orders and customers, but also detailed information about each product and its category.

## Data Transformation

### Orders

```python
import pandas as pd
def transform_orders_df(df: pd.DataFrame) -> pd.DataFrame:
    df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"])
    df["order_purchase_timestamp"] = df["order_purchase_timestamp"].dt.normalize()

    df = df.drop([
        "order_status",
        "order_approved_at",
        "order_delivered_carrier_date", 
        "order_delivered_customer_date", 
        "order_estimated_delivery_date"], axis=1)
    return df
transformed_orders_df = transform_orders_df(orders_df.copy())
```

`order_status` was converted to a categorical `dtype`, and `order_purchase_timestamp` was parsed as datetime ignoring time part because it's not necessary to keep a specific datetime.

Delivery-related attributes such as `order_status`, `order_approved_at`, `order_delivered_carrier_date`, `order_delivered_customer_date` and `order_estimated_delivery_date` were excluded from the initial dataset. Although relevant for logistics analysis, delivery information occurs after the purchase decision and may introduce noise or data leakage in both `Demand Forecasting` and `Customer Segmentation` tasks. Therefore, these variables were intentionally left out in the first iteration.

### Order Items

```python
def transform_order_items_df(df: pd.DataFrame) -> pd.DataFrame:
    df = df.drop(["seller_id", "shipping_limit_date"], axis=1)
    return df
transformed_order_items_df = transform_order_items_df(order_items_df.copy())
```

`seller_id` was discarded because seller shouldn't affect directly the process of purchase so much, so in a first iteration, it would be ideal to avoid any noise data.

`shipping_limit_date` was discarded because it is more closely related to the delivery process rather than the purchase process, which is the focus of the `Demand Forecasting` and `Customer Segmentation` tasks.

### Products

```python
def transform_products_df(df: pd.DataFrame) -> pd.DataFrame:
    df = df.drop(["product_name_lenght",
                     "product_description_lenght",
                     "product_photos_qty",
                     "product_weight_g",
                     "product_length_cm",
                     "product_height_cm",
                     "product_width_cm"], axis=1)
    df["product_category_name"] = df["product_category_name"].fillna("missing")
    return df
transformed_product_df = transform_products_df(products_df.copy())
```

`product_name_lenght`, `product_description_lenght` and `product_photos_qty` are more closely related to user experience. Therefore, in the  first iteration of the project, these features were excluded to reduce noise and keep the model focused on `Demand Forecasting` and `Customer Segmentation`.

According to the Kaggle dataset documentation, `freight_value` already accounts for product weight and dimensions. As a result, including `product_weight_g`, `product_length_cm`, `product_height_cm` and `product_width_cm` would be redundant, so these features were excluded.

A small number of missing values (610 records) were found in `product_category_name`. Instead of dropping these observations, the missing values were filled with a new category labeled `missing`.

### Customers

```python
def transform_customers_df(df: pd.DataFrame) -> pd.DataFrame:
    df = df.drop("customer_zip_code_prefix", axis=1)
    df["customer_city"] = pd.Categorical(df["customer_city"])
    df["customer_state"] = pd.Categorical(df["customer_state"])
    return df
transformed_customer_df = transform_customers_df(customers_df.copy())
```

`customer_zip_code_prefix` was excluded because it provides similar geographic information to `customer_city` and `customer_state`.

`customer_city` and `customer_state` were parsed as categorical `dtype`.

## Data Merging

```python
import pandas as pd
merge_products_product_category_name_translation = pd.merge(transformed_product_df, product_category_name_translation_df, on="product_category_name", how="left")
merge_order_items_products = pd.merge(transformed_order_items_df, merge_products_product_category_name_translation, on="product_id", how="left")
merge_orders_customers = pd.merge(transformed_orders_df, transformed_customer_df, on="customer_id", how="left")
merge_orders_order_items = pd.merge(merge_order_items_products, merge_orders_customers, on="order_id", how="left")
```

Tables were merged to obtain complete order item information, including product and customer data. An left join was used to preserve all order items.

## Post-Merge Validations

```python
olist_df_tmp = merge_orders_order_items.copy()
olist_df_tmp["product_category_name_english"] = olist_df_tmp["product_category_name_english"].fillna("missing")
olist_df_tmp["product_category_name_english"] = pd.Categorical(olist_df_tmp["product_category_name_english"])
olist_df_tmp["product_id"] = pd.Categorical(olist_df_tmp["product_id"])
olist_df_tmp["customer_unique_id"] = pd.Categorical(olist_df_tmp["customer_unique_id"])
olist_df_tmp = olist_df_tmp.dropna(subset=["customer_unique_id"])
```

`olist_df_tmp` was created to fill and transform  `product_category_name_english`, `product_id` and `customer_unique_id` were converted to categorical `dtype` as they are the unique identifiers for products and customers.

15 records with missing values in `customer_unique_id` were discarded.

```python
olist_df_tmp = olist_df_tmp.drop(["customer_id", "product_category_name"], axis=1)
olist_df_tmp = olist_df_tmp.rename(columns={
    'price': 'product_price', 
    'freight_value': 'shipping_cost', 
    'product_category_name_english': 'product_category_name',
    'customer_unique_id': 'customer_id'})
olist_df_tmp.columns
```

Then, `price` was renamed to `product_price` to clarify that it represents the direct price of the product, `freight_value` was renamed to `shipping_cost` to indicate that it corresponeds to the shipping cost and `product_category_name_english` was renamed to `product_category_name` to simplify the column name.

`customer_id` is potentially confusing because it refers to  an order-level identifier, while `customer_unique_id` represent the actual unique customer acrosss multiple orders, for this reason `customer_id` was discarded and `customer_unique_id` was renamed to `customer_id`.

```python
olist_df_tmp = olist_df_tmp.drop(['order_id', 'order_item_id'], axis=1)
print(olist_df_tmp.info())
print(olist_df_tmp.isna().sum())
print(olist_df_tmp.head())
```

Some identifiers were discarded as they do not provide predictive value.
